```python
# matplotlib example work flow

# 1. prepare some data
x = [1,2,3,4,5]
y = [15,12,20,7,20]


# 2. Setup plot
fig , ax = plt.subplots(figsize=(20,12))

# 3. Plot the data
ax.plot(x,y)

# 4. Coustomize the plot
ax.set(title="Title",
       xlabel="x-Axis",
       ylabel="y-Axiz")
```

## PLoting from Pandas DF

```python
car_sales = pd.read_csv("./sample_data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),index=pd.date_range('1/1/2020',periods=1000))

ts = ts.cumsum()

ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
```

```python
car_sales["Price"]
# Converted the price column in valid string
# lets convert it to int
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales["Price"] = car_sales["Price"]/100
```

```python
# Adding the Dates to it

car_sales["Sale Date"] = pd.date_range('1/1/2020',periods=len(car_sales))
car_sales
```

```python
# Added Cumulative sum for price
car_sales["Cumulative-Sum-Price"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Lets plot It!

car_sales.plot(x="Sale Date",y="Cumulative-Sum-Price")
```

```python
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter")
```

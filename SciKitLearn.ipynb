## 1. Scikit-learn Workflow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Getting the Data

```python

heart_disease = pd.read_csv('./sample_data/heart-disease.csv')
heart_disease
```

```python
# --> Creating X (features Column)  and  Y(labels)

X = heart_disease.drop("target", axis=1)

y = heart_disease["target"]
```

## 3. Selecting the right model

```python
# Using the random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()

# We will keep the default parameters
clf.get_params()  # These are hyper parameters
```

## 4. Fitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)  # test_size is 20%
```

```python
clf.fit(X_train, y_train)
```

```python
X_test
```

```python
# make a prediction
y_pred = clf.predict(X_test)
y_pred
```

## 5. Evaluating the Data

```python
clf.score(X_train, y_train)  # it will work perfect
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# --> Classification report
print(classification_report(y_pred=y_pred, y_true=y_test))
```

```python
# --> Confusion matrix
print(confusion_matrix(y_true=y_test, y_pred=y_pred))
```

### 5. Improve the model

```python
# Trying with different hyperparameters
best_n_estimator = 0
current_best_score = 0
for i in range(10, 100, 1):
    print(f"Trying thr model with {i}th estimator")
    clf = RandomForestClassifier(
        n_estimators=i
    )
    clf.fit(X_train, y_train)
    print(f"The Accuracy = {clf.score(X_test,y_test)}")
    if (clf.score(X_test, y_test) > current_best_score):
        current_best_score = clf.score(X_test, y_test)
        best_n_estimator = i

print(
    f"Best n_estimator is {best_n_estimator} with {current_best_score*100}% accuracy")
```

### 6. Saving the model

```python
import pickle

pickle.dump(clf, open("first_model_random_forest.pkl", "wb"))
```

```python
#  load the model

loaded_model = pickle.load(open("./first_model_random_forest.pkl", "rb"))

loaded_model.score(X_test, y_test)
```

### 2.1 Picking a Regression problem
Using sklearn's california dataset (these are real datasets provided by sklearn for examples)

```python
from sklearn.datasets import fetch_california_housing
```

```python
california_dataset = fetch_california_housing()
california_dataset
```

```python
housing_df = pd.DataFrame(california_dataset["data"], columns=california_dataset["feature_names"])
housing_targets = california_dataset["target"]
housing_df["target"]=housing_targets
housing_df
```

```python
# Import the model
# following the estimator map we came to conclosion that we are predicting a value and (considering all the features are equally important) we use RIDGE

from sklearn.linear_model import Ridge

# Setup the random seed
np.random.seed(42)


# features and target
X = housing_df.drop("target",axis=1)
y = housing_df["target"]


# split the Data
X_train,  X_test, y_train,y_test= train_test_split(X,y,test_size=0.3)
```

```python
model = Ridge()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
# it returns the "Co-effecient of determinitation"  
# --> by default it is R^2  ==> it is nothing but the cost function of the linear regration(sum of square of errors)
```
